22. ### Deep Copy vs. Shallow Copy

    - **Shallow Copy** (`copy.copy()`): Copies the top-level object and references to nested objects. Changes to nested objects affect both original and copy.
    ```python
    import copy
    original_list = [1, [2, 3]]
    shallow_copied_list = copy.copy(original_list)
    ```

    - **Deep Copy** (`copy.deepcopy()`): Creates a new object and recursively copies all nested objects. Changes to the copy do not affect the original.
    ```python
    import copy
    original_list = [1, [2, 3]]
    deep_copied_list = copy.deepcopy(original_list)
    ```

    - **Assignment** (`list = list1`): Creates a new reference to `list1`, making both references point to the same object. Changes affect both.
    ```python
    list1 = [1, [2, 3]]
    list = list1
    ```

    - Lists: You can directly reverse a segment with `s[i:i + k] = reversed(s[i:i + k])` since lists are mutable.

    - list.reverse() is useful when you want to reverse a list and don't need to keep the original order.

````python
import re
def average_numbers_in_string(input_string):
    # Extract all numbers (including decimals) from the input string
    numbers = re.findall(r'-?\d+\.?\d*', input_string)
    
    numbers = [float(num) for num in numbers]
    
    '''    
    numbers = re.findall(r'-?\d+', input_string)
    numbers = [int(num) for num in numbers]
    '''

    if numbers:
        return sum(numbers) / len(numbers)
    else:
        return None  # Return None if no numbers found

# Example usage
input_str = "The temperature today is 25.5C, yesterday it was 23C, and tomorrow it might be 27C."
result = average_numbers_in_string(input_str)
print(f"The average of the numbers in the string is: {result}")

def calculate_average_duration(input_string):
    # Split the input string into lines
    lines = input_string.strip().split('\n')
    
    total_duration = 0
    num_days = 0
    
    for line in lines:
        # Split each line into day and duration
        parts = line.split(':')
        if len(parts) == 2:
            # Extract the duration and convert to float
            duration = float(parts[1].strip())
            total_duration += duration
            num_days += 1
    
    # Calculate the average
    if num_days > 0:
        average_duration = total_duration / num_days
        return average_duration
    else:
        return 0

# Example usage
input_data = """Day 1 excercise duration : 25
Day 2 excercise duration: 26
Day 3 excercise duration: 23
Day 4 excercise duration: 15
Day 5 excercise duration: 14
Day 6 excercise duration: 38
Day 7 excercise duration: 44"""

average = calculate_average_duration(input_data)
print(f"The average exercise duration is: {average:.2f}")
````

````python
#O(n)
def is_palindrome(num):
    return str(num) == str(num)[::-1]

def find_palindromes_on(start, end):
    palindromes = []
    for num in range(start, end + 1):
        if is_palindrome(num):
            palindromes.append(num)
    return palindromes
#O(sqrt(n))
def generate_palindromes(start, end):
    def generate_next_palindrome(n):
        s = str(n)
        length = len(s)
        left_half = s[:(length + 1) // 2]
        
        if int(left_half + left_half[-2 if length % 2 == 0 else -1::-1]) > n:
            return int(left_half + left_half[-2 if length % 2 == 0 else -1::-1])
        
        left_half = str(int(left_half) + 1)
        return int(left_half + left_half[-2 if length % 2 == 0 else -1::-1])

    palindrome = generate_next_palindrome(start - 1)
    while palindrome <= end:
        yield palindrome
        palindrome = generate_next_palindrome(palindrome)
````

Certainly! Here's a short and concise explanation of time complexity:

### Time Complexity (Short Version)

1. **Definition**: Time complexity measures how the running time of an algorithm grows with the size of the input, denoted as \(n\).

2. **Big O Notation**: 
   - Written as \(O(f(n))\), it describes the upper limit of an algorithm's run time.
   - Common complexities: 
     - \(O(1)\): Constant time 
     - \(O(n)\): Linear time 
     - \(O(\log n)\): Logarithmic time 

3. **Combining Complexities**:
   - When you have multiple parts, add their complexities, but focus on the dominant term.
   - Example: \(O(\log n) + O(n) = O(n)\) because \(O(n)\) grows faster than \(O(\log n)\) for large \(n\).

4. **Conclusion**: Time complexity helps assess the efficiency of algorithms, allowing comparisons based on how they scale with larger inputs.

In competitive programming, solutions with acceptable time complexity generally depend on the constraints provided in the problem. For a problem like finding the number of ways to climb a staircase with 234 steps, here are the complexities that are usually acceptable:
1. **O(n) Time Complexity**:
    - **Dynamic Programming Approach**: This approach is usually acceptable and efficient for large values of \( n \). For example, in the staircase problem, using dynamic programming or a sliding window technique to calculate the number of ways to climb the stairs is efficient.
    - **Iterative Approach with Constant Space**: Using a sliding window approach to keep track of only the last two computed values is also efficient and acceptable.
    
    ```python
    def climbStairs(n):
        if n == 0 or n == 1:
            return 1
        
        prev2, prev1 = 1, 1
        
        for i in range(2, n + 1):
            current = prev1 + prev2
            prev2 = prev1
            prev1 = current
        
        return prev1

    # Number of steps
    n = 234
    print(climbStairs(n))
    ```

2. **O(log n) Time Complexity**:
    - **Matrix Exponentiation**: For problems related to Fibonacci numbers, matrix exponentiation can be used to achieve O(log n) time complexity, which is even faster than O(n). This is acceptable but more complex to implement.
    
    ```python
    def matrix_mult(A, B):
        return [
            [A[0][0]*B[0][0] + A[0][1]*B[1][0], A[0][0]*B[0][1] + A[0][1]*B[1][1]],
            [A[1][0]*B[0][0] + A[1][1]*B[1][0], A[1][0]*B[0][1] + A[1][1]*B[1][1]]
        ]

    def matrix_pow(mat, exp):
        result = [[1, 0], [0, 1]]
        base = mat

        while exp > 0:
            if exp % 2 == 1:
                result = matrix_mult(result, base)
            base = matrix_mult(base, base)
            exp //= 2
        
        return result

    def climbStairs(n):
        if n == 0 or n == 1:
            return 1
        
        transformation_matrix = [[1, 1], [1, 0]]
        result_matrix = matrix_pow(transformation_matrix, n - 1)
        
        return result_matrix[0][0] + result_matrix[0][1]

    # Number of steps
    n = 234
    print(climbStairs(n))
    ```
Unacceptable Time Complexities

1. **O(2^n) Time Complexity**:
    - **Exponential Recursion**: This approach is generally not acceptable for large values of \( n \) in competitive programming due to its inefficiency. It quickly becomes infeasible as \( n \) increases beyond small values (e.g., \( n > 30 \)).

    ```python
    def climbStairs(n):
        if n == 0 or n == 1:
            return 1
        return climbStairs(n-1) + climbStairs(n-2)

    # Number of steps
    n = 234
    print(climbStairs(n))  # This will take an impractically long time to compute
    ```

**Conclusion** :
For the staircase problem with 234 steps, using either the iterative approach with O(n) time complexity or the matrix exponentiation approach with O(log n) time complexity is acceptable and efficient for competitive programming. The exponential time complexity approach (O(2^n)) is not suitable for large inputs and should be avoided.

````python
def is_palindrome(s):        # time:O(n) space:O(1)
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True
def is_palindrome_slicing(arr): # time:O(n) space:O(n)
    if arr==arr[::-1]:
        return True
    else:
        return False
````

````python
# Method 1: Using a Loop
def reverse_string_loop(s):
    """
    Reverses the string using a loop.
    Time Complexity: O(n^2) - Each iteration involves concatenating the character to the front of the string.
    Space Complexity: O(n) - Creates a new string that grows with the size of the input.
    """
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str
    return reversed_str

# Method 2: Using join and reversed
def reverse_string_join(s):
    """
    Reverses the string using join and reversed.
    Time Complexity: O(n) - Creates a reversed iterator and joins the characters in one pass.
    Space Complexity: O(n) - Creates a new string for the reversed result.
    """
    return "".join(char for char in reversed(s))

# Example usage
print(reverse_string_loop("hello"))  # Output: "olleh"
print(reverse_string_join("hello"))  # Output: "olleh"
````

Excellent question. The reason we only need to check factors up to the square root of n is based on a fundamental property of factors. Let's break this down:

1. Factor pairs: Factors always come in pairs. If a number n has a factor a, it must also have a corresponding factor b, such that a * b = n.

2. Relationship between factor pairs: In any factor pair (a, b) of a number n, if a is less than n, then b must be greater than n (except when a = b = n).

3. Midpoint property: n acts as a midpoint for factor pairs. Once you've checked all numbers up to n, you've implicitly checked all potential factors.

Let's use an example to illustrate:

Consider n = 100

The factors of 100 are: 1, 2, 4, 5, 10, 20, 25, 50, 100

The square root of 100 is 10.

Notice how the factors pair up:
1 and 100
2 and 50
4 and 25
5 and 20
10 and 10

In each pair, at least one factor is less than or equal to 100 (10).

By checking only up to n:
1. We've found all the smaller factors.
2. We've implicitly accounted for all the larger factors (as they must pair with smaller ones).
3. We've avoided redundant checks.

This principle significantly reduces the number of iterations needed:
- For n = 100, we only need to check up to 10 instead of 50.
- For n = 1,000,000, we only check up to 1,000 instead of 500,000.

This optimization becomes increasingly important for larger numbers, dramatically improving the efficiency of algorithms like primality tests or factor finding.

Certainly. Let's dive into the concept of integer overflow and why the formulation `m = l + ((r - l) // 2)` helps avoid it.



Integer overflow occurs when an arithmetic operation attempts to create a numeric value that is outside of the range that can be represented with a given number of bits. In many programming languages, integers have a fixed size (often 32 or 64 bits), which limits the range of values they can represent.

In the context of finding the midpoint in a binary search, the naive formula would be:

```python
m = (l + r) // 2
```

However, this can lead to integer overflow if `l + r` exceeds the maximum value that can be stored in an integer. This is particularly problematic when dealing with very large numbers.

The formula `m = l + ((r - l) // 2)` avoids this problem. Here's why:

1. It first calculates the difference between `r` and `l`, which is guaranteed to be no larger than the original range.
2. It then divides this difference by 2, which further reduces the size.
3. Finally, it adds this result to `l`.

This approach ensures that no intermediate calculation exceeds the original range of values, thus preventing integer overflow.

Let's look at an example:

Suppose we're using 32-bit integers, which typically have a maximum value of 2^31 - 1 = 2,147,483,647.

If `l = 2,000,000,000` and `r = 2,147,483,640`:

1. Using `(l + r) // 2`:
   - `l + r = 4,147,483,640`, which exceeds the maximum 32-bit integer value.
   - This would cause an overflow, leading to incorrect results.

2. Using `l + ((r - l) // 2)`:
   - `r - l = 147,483,640`
   - `(r - l) // 2 = 73,741,820`
   - `l + 73,741,820 = 2,073,741,820`
   - This result is correct and within the range of a 32-bit integer.

By using this formulation, the code remains correct even when dealing with very large integer values, close to the limits of the integer type being used.

````python
def is_perfect_number(n):
    # Time complexity: O(n), Space complexity: O(1)
    if n <= 1:
        return False

    sum_of_divisors = 1  # Start with 1, since 1 is always a divisor of n

    # Check for divisors up to n/2
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            sum_of_divisors += i

    return sum_of_divisors == n

# Example usage
print(is_perfect_number(6))   # Output: True
print(is_perfect_number(28))  # Output: True
print(is_perfect_number(12))  # Output: False
````

When searching for a target in an unsorted array, the most straightforward method is **linear search**, as no sorting has been applied to the array.

**Sorting + Binary Search**:
  - **Sorting Time Complexity**: \(O(n \log n)\)--we can use merge sort
  - **Binary Search Time Complexity**: \(O(\log n)\)
  - **Combined Time Complexity**: \(O(n \log n)\)
    - First, you would sort the array (which takes \(O(n \log n)\)) and then apply binary search to find the target (which takes \(O(\log n)\)).
  


````python
 
def binarySearch(array, value):
    start = 0
    end = len(array)-1
    middle =  (start+end)//2
    while not(array[middle]==value) and start<=end:
        if value < array[middle]:
            end = middle - 1
        else:
            start = middle + 1 
        middle =  (start+end)//2
        # print(start, middle, end)
    if array[middle] == value:
        return middle
    else:
        return -1
custArray = [1, 3, 5, 6, 6, 6, 7, 8, 9]
print(binarySearch(custArray, 6))


def binary_search_first_occurrence(arr, target):
    left = 0
    right = len(arr) - 1
    result = -1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            result = mid  # Potential first occurrence
            right = mid - 1  # Continue searching in the left half
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return result

# Example usage:
sorted_array = [1, 3, 5, 6, 6, 6, 7, 8, 9]
target = 6

result = binary_search_first_occurrence(sorted_array, target)
if result != -1:
    print(f"First occurrence of {target} is at index {result}")
else:
    print(f"Element {target} is not present in the array")
````

````python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
    
    #  Creation of Doubly Linked List
    def createDLL(self, nodeValue):
        node = Node(nodeValue)
        node.prev = None
        node.next = None
        self.head = node
        self.tail = node
        return "The DLL is created Successfully"
    
    
    
    #  Insertion Method in Doubly Linked List
    def insertNode(self, nodeValue, location):
        if self.head is None:
            print("The node cannot be inserted")
        else:
            newNode = Node(nodeValue)
            if location == 0:
                newNode.prev = None
                newNode.next = self.head
                self.head.prev = newNode
                self.head = newNode
            elif location == 1:
                newNode.next = None
                newNode.prev = self.tail
                self.tail.next = newNode
                self.tail = newNode
            else:
                tempNode = self.head
                index = 0
                while index < location - 1:
                    tempNode = tempNode.next
                    index += 1
                newNode.next = tempNode.next
                newNode.prev = tempNode
                newNode.next.prev = newNode
                tempNode.next = newNode
    
    #  Traversal Method in Doubly Linked List
    def traverseDLL(self):
        if self.head is None:
            print("There is not any element to traverse")
        else:
            tempNode = self.head
            while tempNode:
                print(tempNode.value)
                tempNode = tempNode.next
    
    #  Reverse Traversal Method in Doubly Linked List
    def reverseTraversalDLL(self):
        if self.head is None:
            print("There is not any element to traverse")
        else:
            tempNode = self.tail
            while tempNode:
                print(tempNode.value)
                tempNode = tempNode.prev

    # Search Method in Doubly Linked List
    def searchDLL(self, nodeValue):
        if self.head is None:
            return "There is not any element in the list"
        else:
            tempNode = self.head
            while tempNode:
                if tempNode.value == nodeValue:
                    return tempNode.value
                tempNode = tempNode.next
            return "The node does not exist in this list"

    # Delete a node from Doubly Linked List
    def deleteNode(self,location):
        if self.head is None:
            print("There is not any element in DLL")
        else:
            if location == 0:
                if self.head == self.tail:
                    self.head = None
                    self.tail = None
                else:
                    self.head = self.head.next
                    self.head.prev = None
            elif location == 1:
                if self.head == self.tail:
                    self.head = None
                    self.tail = None
                else:
                    self.tail = self.tail.prev
                    self.tail.next = None
            else:
                curNode = self.head
                index = 0
                while index < location - 1:
                    curNode = curNode.next
                    index += 1
                curNode.next = curNode.next.next
                curNode.next.prev = curNode
            print("The node has been successfully deleted")

    # Delete entire Doubly Linked List
    def deleteDLL(self):
        if self.head is None:
            print("There is not any node in DLL")
        else:
            tempNode = self.head
            while tempNode:
                tempNode.prev = None
                tempNode = tempNode.next
            self.head = None
            self.tail = None
            print("The DLL has been successfully deleted")
    


doubyLL = DoublyLinkedList()
doubyLL.createDLL(5)
doubyLL.insertNode(0,0)
doubyLL.insertNode(2,1)
doubyLL.insertNode(6,2)
print([node.value for node in doubyLL]) 
doubyLL.deleteDLL()
print([node.value for node in doubyLL]) 
````

````python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
    
    def __str__(self):
        return str(self.value)

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next
            if node == self.tail.next:
                break
    

class Queue:
    def __init__(self):
        self.linkedList = LinkedList()
    def __str__(self):
        values = [str(x) for x in self.linkedList]
        return ' '.join(values)
    
    def enqueue(self, value):
        newNode = Node(value)
        if self.linkedList.head == None:
            self.linkedList.head = newNode
            self.linkedList.tail = newNode
        else:
            self.linkedList.tail.next = newNode
            self.linkedList.tail = newNode
    
    def isEmpty(self):
        if self.linkedList.head == None:
            return True
        else:
            return False
    
    def dequeue(self):
        if self.isEmpty():
            return "There is not any node in the Queue"
        else:
            tempNode = self.linkedList.head
            if self.linkedList.head == self.linkedList.tail:
                self.linkedList.head = None
                self.linkedList.tail = None
            else:
                self.linkedList.head = self.linkedList.head.next
            return tempNode
    
    def peek(self):
        if self.isEmpty():
            return "There is not any node in the Queue"
        else:
            return self.linkedList.head
    
    def delete(self):
        self.linkedList.head = None
        self.linkedList.tail = None
````

### 3. **Priority Queue**
   - **Definition**: A queue where each element has a priority. Elements with higher priority are dequeued first, regardless of their order of arrival.
   - **Data Structure**: Binary Heap or Balanced Binary Search Tree (e.g., AVL Tree).

---

### 4. **Double-Ended Queue (Deque)**
   - **Definition**: A queue where elements can be added or removed from both ends (front and rear).
   - **Data Structure**: Array or Doubly Linked List.

---

### 5. **Input-Restricted Deque**
   - **Definition**: A type of deque where elements can only be added at one end but removed from both ends.
   - **Data Structure**: Array or Doubly Linked List.

---

### 6. **Output-Restricted Deque**
   - **Definition**: A type of deque where elements can be added at both ends but removed only from one end.
   - **Data Structure**: Array or Doubly Linked List.

````python
class Node:
    def __init__(self, value = None):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def __iter__(self):
        curNode = self.head
        while curNode:
            yield curNode
            curNode = curNode.next

class Stack:
    def __init__(self):
        self.LinkedList = LinkedList()
    
    def __str__(self):
        values = [str(x.value) for x in self.LinkedList]
        return '\n'.join(values)
    
    def isEmpty(self):
        if self.LinkedList.head == None:
            return True
        else:
            return False

    def push(self, value):
        node = Node(value)
        node.next = self.LinkedList.head
        self.LinkedList.head = node
    
    def pop(self):
        if self.isEmpty():
            return "There is not any element in the stack"
        else:
            nodeValue = self.LinkedList.head.value
            self.LinkedList.head = self.LinkedList.head.next
            return nodeValue
    
    def peek(self):
        if self.isEmpty():
            return "There is not any element in the stack"
        else:
            nodeValue = self.LinkedList.head.value
            return nodeValue
    
    def delete(self):
        self.LinkedList.head = None
    



customStack = Stack()
customStack.push(1)
customStack.push(2)
customStack.push(3)
print(customStack)

print(customStack.peek())
````

````python
class MultiStack:

    def __init__(self, stacksize):
        self.numstacks = 5
        self.array = [0] * (stacksize * self.numstacks)
        self.sizes = [0] * self.numstacks
        self.stacksize = stacksize
        # print(self.array)
        # print(self.sizes)

    def Push(self, item, stacknum):
        if self.IsFull(stacknum):
            raise Exception('Stack is full')
        self.sizes[stacknum] += 1
        self.array[self.IndexOfTop(stacknum)] = item

    def Pop(self, stacknum):
        if self.IsEmpty(stacknum):
            raise Exception('Stack is empty')
        value = self.array[self.IndexOfTop(stacknum)]
        self.array[self.IndexOfTop(stacknum)] = 0
        self.sizes[stacknum] -= 1
        return value

    def Peek(self, stacknum):
        if self.IsEmpty(stacknum):
            raise Exception('Stack is empty')
        return self.array[self.IndexOfTop(stacknum)]

    def IsEmpty(self, stacknum):
        return self.sizes[stacknum] == 0

    def IsFull(self, stacknum):
        return self.sizes[stacknum] == self.stacksize

    def IndexOfTop(self, stacknum):
        offset = stacknum * self.stacksize
        return offset + self.sizes[stacknum] - 1

stack = MultiStack(2)
stack.Push(10,1)
stack.Push(9,2)
stack.Push(2,3)
stack.Push(7,2)
stack.Pop(2)
````

Complete Tree: All levels are fully filled except possibly the last level, which is filled from left to right. This tree is always binary, and it can be unbalanced but not sparse.

### Heap and Its Types

A **Heap** is a specialized tree-based data structure that satisfies the **heap property**:
- **Max-Heap**: The value of a parent node is greater than or equal to the values of its children. The largest value is at the root.
- **Min-Heap**: The value of a parent node is less than or equal to the values of its children. The smallest value is at the root.

---

### Types of Heap
1. **Max-Heap**: Used when the maximum value is needed quickly (e.g., priority queues for maximum).
2. **Min-Heap**: Used when the minimum value is needed quickly (e.g., Dijkstras algorithm).

---

### Simple Python Code for a Min-Heap

Heres a simple implementation of a Min-Heap:

```python
import heapq

# Create a Min-Heap
min_heap = []

# Add elements to the heap
heapq.heappush(min_heap, 10)
heapq.heappush(min_heap, 5)
heapq.heappush(min_heap, 20)

# Remove and return the smallest element
smallest = heapq.heappop(min_heap)

# Check the heap
print("Smallest element:", smallest)  # Output: 5
print("Heap after pop:", min_heap)    # Output: [10, 20]
```

---

### Simple Python Code for a Max-Heap

In Python, a **Max-Heap** can be implemented using `heapq` by negating the values:

```python
import heapq

# Create a Max-Heap
max_heap = []

# Add elements to the heap (negate the values)
heapq.heappush(max_heap, -10)
heapq.heappush(max_heap, -5)
heapq.heappush(max_heap, -20)

# Remove and return the largest element (negate back to get the original value)
largest = -heapq.heappop(max_heap)

# Check the heap
print("Largest element:", largest)  # Output: 10
print("Heap after pop:", [-x for x in max_heap])  # Output: [5, 20]
```

---

### Key Points
- **Min-Heap**: Directly supported by Python's `heapq`.
- **Max-Heap**: Simulated using negation (`-value`) because Python's `heapq` defaults to a Min-Heap.

These implementations are simple and use Pythons built-in library, `heapq`, for efficiency and ease.
